```python
import io
import pandas as pd
import numpy as np

with io.open('danny_1008_data.csv', 'r', encoding='latin-1') as file:
    df = pd.read_csv(file)
    df = df[df["LysisMethod"] != "Atila"].reset_index(drop=True)


df.columns
```

```python
df
```

```python
st = df["LysisMethod"][0]

i = 0
def parse_LM(st):
    global i
    print(i, st)
    i += 1


    temp = st.split("1.33x NB")
    op3 = "1.33x NB"
    if len(temp) == 1:
        temp = st.split("water")
        op3 = "water"
        print("===========")
        print(temp)
        
    
    if temp[0] == 'water + ':
        op1, op2 = "water", "-"
    
    elif temp[0] != "":
        op1, op2 = temp[0].split(" ")[:2]     
        
    else:
        op1, op2 = "no", "-"
    
    #print("\t\t", temp)
    op_, op__ = temp[1].split(", ")[1:]
    #print(op_, op__)

    op4, op5 = op_.split(" ")
    op6, op7 = op__.split(" ")[1 if len(op__.split(" "))==3 else 0:]
    print("\t", [op1, op2, op3, op4, op5, op6, op7])
    return [op1, op2, op3, op4, op5, op6, op7]


re = df["LysisMethod"].map(parse_LM)
```

```python
re
```

```python
for i in range(len(re)):
    df.loc[i, ["Concentration", "Enzyme", "Buffer", "LysisTemp", "LysisTime", "InactTemp", "InactTime"]] = re[i]
```
